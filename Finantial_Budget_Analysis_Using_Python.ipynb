```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/expenditure-budget2021-22.xlsx---expenditure-budget-final-file.csv')
df.head()
```

```python
df['Ministries/Departments'].unique()
```

```python
df.drop(['Heads of Expenditure','Actual 2019-2020','Revised 2020-2021'],axis=1)

```

# **Which Ministries/Departments has Highest budget Allotation**

```python
df1=pd.DataFrame()
df1['Ministries/Departments']=df['Ministries/Departments']
df1['Detailed Heads of Expenditure']=df['Detailed Heads of Expenditure']
df1['Budget 2020-2021']=df['Budget 2020-2021']
df1
```

```python
df1.dtypes
```

```python
df1.columns=['Ministries/Departments','Detailed Heads of Expenditure','budget']
df1.head()
```

```python
df1['budget'].replace({'...':0},inplace=True)
df1.head(60)
```

```python
df1['budget']=pd.to_numeric(df1['budget'])
df1.head(62)
```

```python
df1.plot.bar(x='Ministries/Departments',y='budget',figsize=(35,20))
```

```python
df1.dtypes
```

```python
df2=df1.sort_values('budget',ascending=False)
df2
```

```python
highestbudget=df2.head(10)
highestbudget
```

```python
highestbudget.plot(kind='bar',figsize=(30,10),fontsize=20)

plt.xlabel('budget')
plt.ylabel('rupey in caroes')
```

```python
new_df2=df2.groupby('Ministries/Departments').sum().sort_values('budget',ascending=False).reset_index()
new_df2
```

# **distribution of the budget for top 10 ministry**

```python
new_data=new_df2.head(10)
df3=new_data['budget']
labels=new_data['Ministries/Departments']
plt.figure(figsize=(10,9))
plt.pie(df3,labels=labels,autopct='%1.1f%%',startangle=0,pctdistance=0.7,shadow=True,)
central_circle=plt.Circle((0,0),0.1,color='white')
fig=plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font',size=12)
plt.title('Distribution of The Budget For Last 10 Ministries', fontweight ='bold')
plt.show()
```

```python
new_data.plot.bar(x='Ministries/Departments',y='budget',figsize=(20,7))
plt.title("Distribution of The Budget For Top 10 Ministries", fontweight='bold')
plt.show
```
